# 3.	Which Team had won by maximum runs?

```python
matches[matches.win_by_runs==matches.win_by_runs.max()]['winner']
```

# 5.	Which Team had won by closest Margin (minimum runs)?

```python
matches[matches.win_by_runs==1]['winner'].unique()
```

In the case of Cricket, if the team that’s batting second wins the match by chasing, it’ll be said “Team second won by x number of wickets” and if the team that’s bowling second or batted first wins the match by defending their score, it’ll be said “Team first won by x number of runs”. Hence, the minimum win by runs will always be 0 and the minimum win by wickets will also always be 0 in a tournament since sometimes chasing team or sometimes the team that batted first could win.So, to avoid such error we consider the next lowest number than 0, which is 1.

# 9.	Has Toss-winning helped in winning matches?

```python
matches.toss_winner==matches.winner
```

```python
sns.countplot(matches['toss_winner'] == matches['winner'])
plt.title('Matches won by the toss winner vs the matches won by the non toss winner')
plt.show()
```

Clearly, the matches won by the toss winner are slightly higher by that of the matches won by the team who
did not win the toss. This is just a percent differance and hence we can conclude by saying that the teams have a 50-50 chance in winning or losing.
